2) What kind of API did you build? What types of APIs are there? Why are APIs important?

It depends on what you mean by "type"!  In a broad sense, there are public APIs (that anyone can access, often involving some sort of authentication like an API key), private or internal APIs (only available to users within an organization or organizations), and composite APIs that utilize a combination of public and private APIs.  In terms of protocols, RESTful (representational state transfer) are probably the most popular/common these days.  They are stateless, utilize a client/server approach, and are pretty flexible in the types of things they can do and accomplish - for example, REST APIs can "speak" several coding langauges, including HTML, XML, and JSON.  The more old-fashioned SOAP (simple object access protocol) is more highly structured and less flexible than REST - for example, SOAP APIs are built on XML and don't "speak" JSON.  Since our API utilizes JSON, it can't be SOAP, so I'm guessing it's RESTful.

APIs (aka Application Programming Interfaces) are pretty much the basis for everything on the web these days.  They allow our smartphones, computers, and other internet-connected devices to connect to, communicate and exchange information with remote servers all over the globe.  Social networks like Facebook, apps like Uber, online bank services, and most other networked companies/services you can think of utilize APIs.  APIs allow for dynamic and efficient data retrieval and sharing, which in turn makes for faster software development and implementation (since developers don't have to reinvent the wheel and create software/code from scratch for everything).
