```python
import regex as re
splitter = r"[.?!]"
split_text = re.split(splitter, nltk.corpus.gutenberg.raw('milton-paradise.txt'))
print(split_text)
```

```python
sent_list = []
for sentence in split_text:
    sentiments = (sentence, analyzer.polarity_scores(sentence))
    sent_list.append(sentiments)
    print(sentiments)
```

Depends on your point of view!  The sentiments of the individual words probably haven't changed very much over the past few hundred years and we can rubberstamp the scores as being pretty "correct" (even if we're not using "spake" and "thou" that much these days).  From a historical, religious, or literary standpoint, determining the "correctness" of the poem's meaning is much more complex than a basic sentiment analysis can extract.  If you're C.S. Lewis, you're pro-"Paradise Lost"; if you're Ezra Pound, you think that "Milton is the worst sort of poison." And if you're the German empire in 1732 or the Catholic church in 1758, you're banning "Paradise Lost," while if you're a bookstore in Kuwait in 2013, you're getting shut down for carrying a translation of it.  All that to say: there isn't even close to a human consensus on the work's value, so how can we expect a machine to accurately grade its contents?
