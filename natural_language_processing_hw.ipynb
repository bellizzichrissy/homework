```python
text_sentences = nltk.corpus.gutenberg.raw('milton-paradise.txt')
print(text_sentences)
```

```python
from nltk.tokenize import word_tokenize
text_words = word_tokenize(text_sentences)
text_words
```

```python
lower_s = [w.lower() for w in text_words]
print(lower_words[:50])
```

```python
filtered_words = [word for word in lower_words if word not in stop_words]
filtered_words
```

```python
for sentence in filtered_words:
    new_filtered_list = [word for word in filtered_words if word.isalnum()]

print(new_filtered_list[:50])
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in new_filtered_list])
print(lemmatized_output)
```

```python
lemma_words = word_tokenize(lemmatized_output)
unique_tokens = set(lemma_words)
print(unique_tokens)
```
